```python
'''
Que-2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Ans - The scatter() function in Matplotlib's pyplot module is used to create scatter plots. This function takes two arrays 
      (or lists) of data points, one for the x-axis values and the other for the y-axis values. You can also customize the 
      appearance of the dots by specifying attributes like color, size, and shape.
'''
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y,color='b')
plt.title('This is title')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

```python
'''
Que-3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Ans - The subplot() function in Matplotlib is used to create multiple plots (subplots) within a single figure. 
      It's particularly useful when you want to display multiple plots side by side for comparison or visualization of 
      related data. The subplot() function takes three arguments: the number of rows, the number of columns, and the index 
      of the subplot you're creating.
      
'''
# Data for four line plots
x = np.array([0,1,2,3,4,5])

# Data for each line plots
y1 = np.array([0,100,200,300,400,500])
y2 = np.array([50,20,40,20,60,70])
y3 = np.array([10,20,30,40,50,60])
y4 = np.array([200,350,250,550,450,150])
plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.plot(x,y1,marker='o')
plt.title('Line Plot 1')

plt.subplot(2,2,2)
plt.plot(x,y2,marker='s')
plt.title('Line Plot 2')

plt.subplot(2,2,3)
plt.plot(x,y3,marker='^')
plt.title('Line Plot 3')

plt.subplot(2,2,4)
plt.plot(x,y4,marker='d')
plt.title('Line Plot 4')

plt.tight_layout()
plt.show()
```

```python
'''
Que-4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

Ans - A bar plot, also known as a bar chart, is a type of data visualization that represents data using rectangular bars. 
      Each bar's length or height is proportional to the value it represents. Bar plots are used to compare and display the 
      values of different categories or groups.
      
      Bar graph is used for - 
      1:Displaying data in categorical form 
      2:comparing data
      3:showing trends 
'''
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(8,6))
plt.bar(company,profit,color='blue')
plt.xlabel("Company")
plt.ylabel("Profit $ ")
plt.title("Company Profits ")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.barh(company,profit,color='green')
plt.xlabel("Profit $")
plt.ylabel("Company  ")
plt.title("Company Profits ")
plt.show()
```

```python
'''
Que-5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
Ans - A box plot, also known as a box-and-whisker plot, is a graphical representation that displays the distribution of a 
      dataset along with important statistical information. It provides a visual summary of the minimum, first quartile,
      median, third quartile, and maximum values of a dataset, along with potential outliers.
      It is used for - 
      1- Identifying skewness
      2:Detecting Outliers
      3-Comparing
      4-Visualizing spread 
'''

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1,box2]
plt.figure(figsize=(8,6))
plt.boxplot(data,labels=['box1','box2'])
plt.xlabel("Box")
plt.ylabel("Value")
plt.title("Box plot Example")
plt.show()
```
