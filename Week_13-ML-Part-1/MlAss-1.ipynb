```python
'''
Que-1: Explain the following with an example 
1- Artificial Intelligence
2- Machine Learning
3- Deep Learning 

Ans - Artificial Intelligence (AI):

Artificial Intelligence refers to the development of computer systems that can perform tasks that typically require human 
intelligence. These tasks include things like understanding natural language, recognizing patterns, making decisions, and 
solving problems. AI systems can be rule-based or learned from data.

Example: An AI-powered language translation system, like Google Translate, can understand and translate text from one language
to another. It learns from a vast amount of text data to improve its accuracy over time.

Machine Learning (ML):

Machine Learning is a subset of AI that focuses on developing algorithms and models that allow computers to learn from data and
improve their performance on a specific task without being explicitly programmed. ML algorithms can identify patterns and make 
predictions or decisions based on the data they've been trained on.

Example: Consider an email spam filter. Instead of programming explicit rules for identifying spam, a machine learning model 
can be trained on a dataset of labeled emails (spam or not spam). The model learns to recognize patterns in the emails and can
then classify incoming emails as spam or not based on what it has learned.

Deep Learning:

Deep Learning is a specialized subset of machine learning that involves neural networks with multiple layers (deep neural 
networks). These networks attempt to mimic the human brain's structure to process data and learn patterns. Deep learning has 
been particularly successful in tasks like image and speech recognition.

Example: Image classification is a common application of deep learning. A deep learning model, such as a Convolutional Neural 
Network (CNN), can be trained to recognize objects in images. For instance, a deep learning model trained on a large dataset of
cat images can accurately identify whether a new image contains a cat or not.

'''
```

```python
'''
Que-2- What is supervised learning*? List some examples of supervised learning.
Ans - Supervised learning is a type of machine learning where the algorithm learns from labeled training data. In supervised 
learning, the model is provided with input-output pairs (also known as features and labels), and it learns to map the inputs 
to the correct outputs. The goal of supervised learning is to learn a mapping function from the input data to the corresponding
target outputs so that the model can make accurate predictions on new, unseen data.

Examples of supervised learning tasks:

Image Classification: Given a dataset of images and their corresponding labels, a supervised learning algorithm can be trained
to classify images into predefined categories. For example, classifying images of animals as dogs, cats, or birds.

Spam Email Detection: A model can be trained to distinguish between spam and non-spam emails based on features extracted from 
the email content.

Medical Diagnosis: Using medical data, a model can be trained to predict whether a patient has a particular disease based on 
symptoms, test results, and other relevant factors.

House Price Prediction: Given features like the number of bedrooms, location, and square footage, a model can predict the 
selling price of a house.

'''
```

'''
Que-3- What is unsupervised learning? List some examples of unsupervised learning.

Ans - Unsupervised learning is a type of machine learning where the algorithm learns from unlabeled data, meaning there are no 
predefined output labels. In unsupervised learning, the goal is to discover patterns, relationships, or structures within the 
data without explicit guidance in the form of labeled examples. The algorithm tries to identify natural groupings or clusters 
in the data or reduce its dimensionality.

In unsupervised learning, the process typically involves the following steps:

Data Collection: Gather a dataset containing only input data, without any associated output labels.

Model Training: The algorithm analyzes the patterns and relationships within the data to uncover hidden structures. This might 
involve clustering similar data points together or reducing the data's complexity.

Analysis: After training, the model doesn't produce predictions as in supervised learning. Instead, it provides insights into 
the data's inherent structure, such as identifying clusters or patterns.

Examples of unsupervised learning tasks:

Clustering: Clustering algorithms group similar data points together based on their features. For example, grouping customers 
into segments based on their purchasing behavior without any predefined labels.

Dimensionality Reduction: Techniques like Principal Component Analysis (PCA) reduce the number of features in the data while 
retaining as much useful information as possible. This is useful for visualization and reducing computational complexity.

Anomaly Detection: Unsupervised learning can be used to identify unusual or rare data points in a dataset, such as detecting 
fraudulent transactions or faulty products.

Topic Modeling: Unsupervised methods like Latent Dirichlet Allocation (LDA) can be used to discover topics in a collection of 
documents without knowing the specific topics in advance.

Recommendation Systems: Unsupervised learning can analyze user behavior and preferences to suggest items that users might be
interested in.
